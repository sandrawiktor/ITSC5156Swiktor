```python
import pandas as pd
df = pd.read_json('/content/reviews_Musical_Instruments_5 (1).json', lines=True)

```

```python
df = df[['reviewText', 'overall']]
```

```python
df
```

```python
import nltk
nltk.download('punkt')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
# holds all tokenized reviews
reviews_tok = []
```

```python
tokenizer = nltk.RegexpTokenizer(r"\w+")


for i in df['reviewText']:  
  
  reviews_tok.append(tokenizer.tokenize(i))
```

```python
reviews_tok
```

```python
tag = []
```

```python
for i in reviews_tok:
  tag.append(nltk.pos_tag(i))
```
